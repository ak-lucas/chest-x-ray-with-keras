# Visualização do desempenho no treinamento

```python
import matplotlib.pyplot as plt
import csv

f_name = ['training_fold_1.log',
          'training_fold_2.log',
          'training_fold_3.log',
          'training_fold_4.log',
          'training_fold_5.log']

for i in range(len(f_name)):
    loss = []
    val_loss = []
    acc = []
    val_acc = []
    print "----------------------------------------------------------------------"
    with open('../logs/' + f_name[i]) as f:
        reader = csv.reader(f)
        # pula cabeçalho
        next(reader)
        for r in reader:
            acc.append(float(r[1]))
            loss.append(float(r[2]))
            val_acc.append(float(r[3]))
            val_loss.append(float(r[4]))
    print "\n--LOSS"
    plt.rcParams["figure.figsize"] = ["15", "15"]
    plt.plot(range(1,len(loss)+1), loss, color='black', linewidth=2, label='train_loss')
    plt.plot(range(1,len(val_loss)+1), val_loss, color='red', linewidth=2, label='val_loss')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    rangex = range(1,len(loss)+1, 5)
    plt.xticks(rangex, rotation=0)
    plt.title(f_name[i])
    plt.grid(True)
    leg = plt.legend(loc='best', ncol=2, shadow=True, fancybox=True)
    leg.get_frame().set_alpha(0.5)

    plt.show()
    
    print "\n--ACC"
    plt.plot(range(1,len(acc)+1), acc, color='black', linewidth=2, label='train_acc')
    plt.plot(range(1,len(val_acc)+1), val_acc, color='red', linewidth=2, label='val_acc')
    plt.xlabel('epoch')
    plt.ylabel('acc')
    plt.yticks([0.4, 0.5, 0.6, 0.7] + [max(val_acc)] + [val_acc[-1]])
    plt.xticks(range(1,len(loss)+1, 5), rotation=0)
    plt.title(f_name[i])
    plt.grid(True)
    leg = plt.legend(loc='best', ncol=2, shadow=True, fancybox=True)
    leg.get_frame().set_alpha(0.5)

    plt.show()
    print "----------------------------------------------------------------------"
```
